````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from pathlib import Path

file_path = Path("access_log.csv")

df = pd.read_csv(file_path)
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df = df.sort_values(["Employee_ID", "Timestamp"]).reset_index(drop=True)

df.head()
````

## Töömall analüüsi alustamiseks

Allolev näide on mõeldud esialgse tööversiooni koostamise toetamiseks. Vajadusel kohandage veerunimesid, failinime ja analüüsi loogikat vastavalt oma andmestikule.

1-tunnise meeskonnatöö puhul on soovitatav lahendada ülesanne etapiviisiliselt:
- esmalt andmete puhastamine
- seejärel põhinäitajate arvutamine
- seejärel 2 kuni 3 graafiku koostamine teekidega `seaborn` ja `plotly`
- lõpuks `Gradio` prototüübi vormistamine

````python
events = df.copy()
events["event_delta"] = events["Access_Point"].map({"Sisenemine": 1, "Väljumine": -1})
events["date"] = events["Timestamp"].dt.date

entries = (
    events[events["Access_Point"] == "Sisenemine"]
    .copy()
    .rename(columns={"Timestamp": "entry_time"})
)
entries["visit_number"] = entries.groupby(["Employee_ID", "date"]).cumcount()

exits = (
    events[events["Access_Point"] == "Väljumine"]
    .copy()
    .rename(columns={"Timestamp": "exit_time"})
)
exits["visit_number"] = exits.groupby(["Employee_ID", "date"]).cumcount()

visits = entries.merge(
    exits[["Employee_ID", "date", "visit_number", "exit_time"]],
    on=["Employee_ID", "date", "visit_number"],
    how="left",
)

visits["hours_in_building"] = (
    visits["exit_time"] - visits["entry_time"]
).dt.total_seconds() / 3600

hourly_occupancy = (
    events.sort_values("Timestamp")
    .set_index("Timestamp")["event_delta"]
    .cumsum()
    .resample("1h")
    .max()
    .ffill()
    .fillna(0)
    .rename("people_in_building")
)

longest_stays = (
    visits.groupby("Employee_ID", as_index=False)["hours_in_building"]
    .sum()
    .sort_values("hours_in_building", ascending=False)
)

long_workdays = visits[visits["hours_in_building"] > 10].copy()
unusual_cases = visits[visits["exit_time"].isna()].copy()

department_summary = (
    visits.groupby("Department", as_index=False)["hours_in_building"]
    .sum()
    .sort_values("hours_in_building", ascending=False)
)

longest_stays.head()
````

````python
plt.figure(figsize=(12, 5))
sns.lineplot(x=hourly_occupancy.index, y=hourly_occupancy.values)
plt.title("People in the building by hour")
plt.xlabel("Time")
plt.ylabel("People count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plotly_figure = px.bar(
    department_summary,
    x="Department",
    y="hours_in_building",
    title="Total hours in the building by department",
)
plotly_figure.show()

unusual_cases.head()
````

## Kohustuslik osa: lihtne `Gradio` prototüüp

Ülesande üheks kohustuslikuks osaks on väikese veebirakenduse koostamine. Eesmärk ei ole valmis toote loomine, vaid näidata, et andmeanalüüsi tulemusi on võimalik esitada ka kasutajaliidese kaudu.

Kui `gradio` ei ole paigaldatud, saab selle paigaldada näiteks käsuga `pip install gradio`.

````python
import gradio as gr


def summarize_access_file(uploaded_file):
    """Create a small access-log summary from an uploaded CSV file."""
    data = pd.read_csv(uploaded_file.name)
    data["Timestamp"] = pd.to_datetime(data["Timestamp"])
    data = data.sort_values("Timestamp").reset_index(drop=True)
    data["event_delta"] = data["Access_Point"].map({"Sisenemine": 1, "Väljumine": -1})

    hourly_summary = (
        data.set_index("Timestamp")["event_delta"]
        .cumsum()
        .resample("1h")
        .max()
        .ffill()
        .fillna(0)
        .rename("people_in_building")
        .reset_index()
    )

    return hourly_summary.head(24)


demo = gr.Interface(
    fn=summarize_access_file,
    inputs=gr.File(label="Upload CSV file"),
    outputs=gr.Dataframe(label="Hourly summary"),
    title="Building Access Analyzer",
    description="Upload an access log and preview hourly occupancy.",
)

# demo.launch()
````

## 1. Teegi importimine ja CSV-faili laadimine

````python
import pandas as pd

# Loe CSV-fail sisse
raw_df = pd.read_csv('data/access_log_1000.csv')

# Teisenda Timestamp datetime tüübiks
raw_df['Timestamp'] = pd.to_datetime(raw_df['Timestamp'])

# Pivot nii, et Sisenemine ja Väljumine muutuvad eraldi veergudeks
pivot_df = raw_df.pivot_table(
    index=['Employee_ID', 'Department'],
    columns='Access_Point',
    values='Timestamp',
    aggfunc='first'
).reset_index()

# Nimeta veerud ümber vastavalt oodatavale struktuurile
pivot_df.columns.name = None
df = pivot_df.rename(columns={
    'Employee_ID': 'töötaja_id',
    'Department': 'osakond',
    'Sisenemine': 'sisenemine',
    'Väljumine': 'väljumine'
})

# Lisa veerg "nimi"
df['nimi'] = 'Töötaja ' + df['töötaja_id'].astype(str)

# Arvuta tundide arv
df['tunde_hoones'] = (df['väljumine'] - df['sisenemine']).dt.total_seconds() / 3600

df
````

## 5. Rühmitamine: groupby

`groupby` töötab kolme sammuna:

```
df.groupby('osakond')['tunde_hoones'].sum()
     ↑                    ↑             ↑
  rühmita              vali veerg    arvuta
```

| Funktsioon | Tähendus |
|---|---|
| `.sum()` | Summa |
| `.mean()` | Keskmine |
| `.count()` | Kirjete arv |
| `.max()` / `.min()` | Maksimaalne / minimaalne väärtus |
| `.agg([...])` | Mitu funktsiooni korraga |

````python
# Kogutunnid osakondade kaupa — leiab kõige aktiivsema osakonna
osakond_tunnid = (
    df.dropna(subset=['tunde_hoones'])  # jäta välja anomaaliad
      .groupby('osakond')['tunde_hoones']
      .agg(['sum', 'mean', 'count'])
      .rename(columns={'sum': 'kokku_tundi', 'mean': 'keskmine', 'count': 'kirjeid'})
      .sort_values('kokku_tundi', ascending=False)
)
osakond_tunnid
````

````python
# Top 5 töötajat — kes viibis hoones kõige kauem kokku?
top_tootajad = (
    df.dropna(subset=['tunde_hoones'])
      .groupby(['töötaja_id', 'nimi', 'osakond'])['tunde_hoones']
      .sum()
      .reset_index()           # muuda grupeerimise tulemus tagasi DataFrame'iks
      .sort_values('tunde_hoones', ascending=False)
      .head(5)
)
top_tootajad
````

## 6. Inimeste arv hoones tunni kaupa

See on ülesande keerulisem osa. Peame iga tunni kohta (6–22) loendama,
mitu töötajat sel hetkel hoones viibis.

**Loogika:** töötaja on hoones tunnil `t`, kui:
```
sisenemine_tund <= t  JA  väljumine_tund >= t
```

````python
tunnid = range(6, 23)  # 6:00 kuni 22:00
tulemused = {}

for tund in tunnid:
    mask = (
        (df['sisenemine_tund'] <= tund) &
        (df['väljumine_tund'].fillna(23) >= tund)  # NaN → 23 (eeldame, et on veel sees)
    )
    tulemused[tund] = mask.sum()

# Muuda sõnastik DataFrame'iks
tunnid_df = pd.DataFrame.from_dict(tulemused, orient='index', columns=['inimeste_arv'])
tunnid_df.index.name = 'tund'
tunnid_df = tunnid_df.reset_index()

# Leia tipptund ja tühjeim tund
tipptund  = tunnid_df.loc[tunnid_df['inimeste_arv'].idxmax(), 'tund']
tyhjim    = tunnid_df.loc[tunnid_df['inimeste_arv'].idxmin(), 'tund']

print(f'Tipptund:        {tipptund}:00 ({tunnid_df["inimeste_arv"].max()} inimest)')
print(f'Tühjeim kellaaeg: {tyhjim}:00 ({tunnid_df["inimeste_arv"].min()} inimest)')

tunnid_df
````
