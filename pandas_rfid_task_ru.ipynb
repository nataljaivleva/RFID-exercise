````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from pathlib import Path

file_path = Path("access_log.csv")

df = pd.read_csv(file_path)
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df = df.sort_values(["Employee_ID", "Timestamp"]).reset_index(drop=True)

df.head()
````

## Шаблон для начала анализа

Пример ниже предназначен для подготовки первой рабочей версии решения. При необходимости измените названия столбцов, имя файла и логику анализа в соответствии с Вашим датасетом.

- сначала очистка данных
- затем расчет основных показателей
- затем построение 2-3 графиков с помощью `seaborn` и `plotly`
- в конце оформление прототипа на `Gradio`

````python
events = df.copy()
events["event_delta"] = events["Access_Point"].map({"Sisenemine": 1, "Väljumine": -1})
events["date"] = events["Timestamp"].dt.date

entries = (
    events[events["Access_Point"] == "Sisenemine"]
    .copy()
    .rename(columns={"Timestamp": "entry_time"})
)
entries["visit_number"] = entries.groupby(["Employee_ID", "date"]).cumcount()

exits = (
    events[events["Access_Point"] == "Väljumine"]
    .copy()
    .rename(columns={"Timestamp": "exit_time"})
)
exits["visit_number"] = exits.groupby(["Employee_ID", "date"]).cumcount()

visits = entries.merge(
    exits[["Employee_ID", "date", "visit_number", "exit_time"]],
    on=["Employee_ID", "date", "visit_number"],
    how="left",
)

visits["hours_in_building"] = (
    visits["exit_time"] - visits["entry_time"]
).dt.total_seconds() / 3600

hourly_occupancy = (
    events.sort_values("Timestamp")
    .set_index("Timestamp")["event_delta"]
    .cumsum()
    .resample("1h")
    .max()
    .ffill()
    .fillna(0)
    .rename("people_in_building")
)

longest_stays = (
    visits.groupby("Employee_ID", as_index=False)["hours_in_building"]
    .sum()
    .sort_values("hours_in_building", ascending=False)
)

long_workdays = visits[visits["hours_in_building"] > 10].copy()
unusual_cases = visits[visits["exit_time"].isna()].copy()

department_summary = (
    visits.groupby("Department", as_index=False)["hours_in_building"]
    .sum()
    .sort_values("hours_in_building", ascending=False)
)

longest_stays.head()
````

````python
plt.figure(figsize=(12, 5))
sns.lineplot(x=hourly_occupancy.index, y=hourly_occupancy.values)
plt.title("Количество людей в здании по часам")
plt.xlabel("Время")
plt.ylabel("Количество людей")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plotly_figure = px.bar(
    department_summary,
    x="Department",
    y="hours_in_building",
    title="Суммарное время пребывания в здании по отделам",
)
plotly_figure.show()

unusual_cases.head()
````

## Обязательная часть: простой прототип `Gradio`

Одной из обязательных частей задания является создание небольшого веб-приложения. Цель состоит не в разработке готового продукта, а в том, чтобы показать, как результаты анализа данных можно представить через пользовательский интерфейс.

Если `gradio` не установлен, его можно установить, например, командой `pip install gradio`.

````python
import gradio as gr


def summarize_access_file(uploaded_file):
    """Создает краткую сводку по журналу доступа из загруженного CSV-файла."""
    data = pd.read_csv(uploaded_file.name)
    data["Timestamp"] = pd.to_datetime(data["Timestamp"])
    data = data.sort_values("Timestamp").reset_index(drop=True)
    data["event_delta"] = data["Access_Point"].map({"Sisenemine": 1, "Väljumine": -1})

    hourly_summary = (
        data.set_index("Timestamp")["event_delta"]
        .cumsum()
        .resample("1h")
        .max()
        .ffill()
        .fillna(0)
        .rename("people_in_building")
        .reset_index()
    )

    return hourly_summary.head(24)


demo = gr.Interface(
    fn=summarize_access_file,
    inputs=gr.File(label="Загрузите CSV-файл"),
    outputs=gr.Dataframe(label="Почасовая сводка"),
    title="Анализатор данных системы доступа",
    description="Загрузите журнал доступа и просмотрите почасовую заполненность здания.",
)

# demo.launch()
````

## Что такое pandas и почему у него такое название?

**pandas** - это библиотека Python для обработки данных. Название происходит от английского выражения:

> **pan**el **da**ta → **pandas**

**Panel data** - это термин из экономики, обозначающий данные, в которых **одни и те же объекты наблюдаются в разные моменты времени**. Например, это могут быть события входа сотрудников в здание в разные даты. Именно такие данные мы и анализируем в этом задании.

Библиотека изначально разрабатывалась **Wes McKinney** с 2008 года, сначала для анализа финансовых данных в AQR Capital Management.

## 1. Импорт библиотеки и загрузка CSV-файла

````python
import pandas as pd

# Чтение CSV-файла
raw_df = pd.read_csv('data/access_log_1000.csv')

# Преобразование Timestamp в тип datetime
raw_df['Timestamp'] = pd.to_datetime(raw_df['Timestamp'])

# Pivot так, чтобы Sisenemine и Väljumine стали отдельными столбцами
pivot_df = raw_df.pivot_table(
    index=['Employee_ID', 'Department'],
    columns='Access_Point',
    values='Timestamp',
    aggfunc='first'
).reset_index()

# Переименование столбцов в ожидаемую структуру
pivot_df.columns.name = None
df = pivot_df.rename(columns={
    'Employee_ID': 'töötaja_id',
    'Department': 'osakond',
    'Sisenemine': 'sisenemine',
    'Väljumine': 'väljumine'
})

# Добавление столбца "nimi"
df['nimi'] = 'Töötaja ' + df['töötaja_id'].astype(str)

# Расчет количества часов
df['tunde_hoones'] = (df['väljumine'] - df['sisenemine']).dt.total_seconds() / 3600

df
````

## 5. Группировка: groupby

`groupby` работает в три шага:

```
df.groupby('osakond')['tunde_hoones'].sum()
     ↑                    ↑             ↑
 группируй          выбери столбец   вычисли
```

| Функция | Значение |
|---|---|
| `.sum()` | Сумма |
| `.mean()` | Среднее |
| `.count()` | Количество записей |
| `.max()` / `.min()` | Максимальное / минимальное значение |
| `.agg([...])` | Несколько функций сразу |

````python
# Общее количество часов по отделам - помогает найти самый активный отдел
osakond_tunnid = (
    df.dropna(subset=['tunde_hoones'])  # исключаем аномалии
      .groupby('osakond')['tunde_hoones']
      .agg(['sum', 'mean', 'count'])
      .rename(columns={'sum': 'kokku_tundi', 'mean': 'keskmine', 'count': 'kirjeid'})
      .sort_values('kokku_tundi', ascending=False)
)
osakond_tunnid
````

````python
# Топ-5 сотрудников - кто провел в здании больше всего времени?
top_tootajad = (
    df.dropna(subset=['tunde_hoones'])
      .groupby(['töötaja_id', 'nimi', 'osakond'])['tunde_hoones']
      .sum()
      .reset_index()           # превращаем результат группировки обратно в DataFrame
      .sort_values('tunde_hoones', ascending=False)
      .head(5)
)
top_tootajad
````

## 6. Количество людей в здании по часам

Это более сложная часть задания. Нужно посчитать для каждого часа (6-22),
сколько сотрудников находилось в здании в этот момент.

**Логика:** сотрудник находится в здании в час `t`, если:
```
sisenemine_tund <= t  И  väljumine_tund >= t
```

````python
tunnid = range(6, 23)  # с 6:00 до 22:00
tulemused = {}

for tund in tunnid:
    mask = (
        (df['sisenemine_tund'] <= tund) &
        (df['väljumine_tund'].fillna(23) >= tund)  # NaN -> 23 (предполагаем, что сотрудник еще внутри)
    )
    tulemused[tund] = mask.sum()

# Преобразуем словарь в DataFrame
tunnid_df = pd.DataFrame.from_dict(tulemused, orient='index', columns=['inimeste_arv'])
tunnid_df.index.name = 'tund'
tunnid_df = tunnid_df.reset_index()

# Находим пиковый и самый пустой час
tipptund  = tunnid_df.loc[tunnid_df['inimeste_arv'].idxmax(), 'tund']
tyhjim    = tunnid_df.loc[tunnid_df['inimeste_arv'].idxmin(), 'tund']

print(f'Пиковый час:      {tipptund}:00 ({tunnid_df["inimeste_arv"].max()} человек)')
print(f'Самое пустое время: {tyhjim}:00 ({tunnid_df["inimeste_arv"].min()} человек)')

tunnid_df
````
